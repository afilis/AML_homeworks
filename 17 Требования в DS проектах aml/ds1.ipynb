Возьмите за основу продукт компании “Нетология” - профессию Data Scientist.  
Как бы Вы организовали аналитику по данному продукту?  
Сформулируйте 2-3 ключевые метрики продукта  
Какие данные Вам могут помочь в поиске инсайтов?  
Опишите команду для работы с данными для данного продукта  
Какие роли Вам будут нужны?  
Чем они будут заниматься?  

*Цели и задачи*

Основная задача - обеспечить повышение уровня продаж продукта "профессия Data Scientist".  
Главной целью будет поиск проблем недостаточного уровня продаж и возможностей для их повышения. 

*Метрики*

Можно выделить 4 метрики.  
Основой метрикой будет уровень продаж.  
Метрика востребованности выпускников курса будет отображать качество подготовки специалистов.  
Метрика известности позволит понять объем аудитории, знакомой с продуктом.  
Метрика отношения позволит оценить какое отношение к курсу и компании в целом. 

*Какие данные могут помочь*

1) Текущий уровень продаж по месяцам от отдела продаж. На основе этих данных можно будет выявить сезонные факторы и возможные аномалии.  
2) Получить данные от отдела маркетинга по проводимым рекламным компаниям и проанализировать корреляцию с данными отдела продаж. Также, получить данные по проведенным исследованиям в т.ч. по стоимости продукта и предоставляемым скидкам.  
3) Провести опрос среди выпускников о том, как быстро они смогли получить работу, использовали ли при этом данные от отдела поиска вакансий.  
Эти данные будут крайне полезны, однако, получить их скорее всего будет не просто, т.к. люди как правило не любят прохожить опросы.  
4) Сделать опрос по ряду компаний, о которых известно, что в них работают выпускники, на тему того насколько полезны принятые ими специалисты и чего не хватает в их образовании. Опять же, эти данные будет не просто получить, т.к. компании скорее всего не будут разглашать такую информацию, но она была бы очень полезна для выяснения ситуации.  
5) Провести опрос среди компаний, нанимающих специалистов Data Scientist, о полезности элементов программы курса с целью ее коррекции.  
6) Провести опрос среди студентов курса о том, как они узнали о компании и этом курсе, почему среди прочих компаний выбрали именно эту, насколько организация процесса обучения, уровень получаемых знаний соответствует их ожиданиям, чего не хватает или что можно улучшить в курсе.  
7) Провести опрос среди преподавателей курса о том, насколько выданный ими материал полон и полезен для слушателей, как они оценивают реакцию аудитории и их вовлеченность в процесс. На основе этих данных можно будет понять как можно скорректировать программу курса, чтобы сделать ее более полной и доступной для понимания.  
8) Собрать информацию по отзывам пользователей на различных сайтах (форумах) с отзывами как по Нетологии, так и по другим компаниям, продающим аналогичный продукт. Это позволит выявлять проблемы, с которыми сталкивались студенты курсов в разных компаниях, что поможет в улучшении продукта. Также, данные по другим компаниям позволят правильнее нормализировать данные по Нетологии (уровень настроений может быть негатиынм или позитивным или негативным по всем компаниям).  
9) Собрать данные от поисковых систем как часто запрашивали в поиске курс "профессия Data Scientist" и как часто искали "Нетология". это позволит оценивать известность.  

*Какие специалисты понадобятся*

- BI-аналитик для анализа бизнес-данных от отдела маркетинга, продаж др. Поможет оценить текущее состояние бизнеса и результаты внедрения предложений от DS-специалистов, представить их для бизнеса и методистов.  
- аналитик данных будет заниматься анализом взаимосвязи признаков, проводить А/Б-тесты, отслеживать изменения метрик  
- DS-специалист будет строить модели, позволяющие предсказывать изменения метрик, формулировать гипотезы, искать решения  
- разработчик будет создавать решения для проведения опросов, средств организации и проведения занятий, средств продвижения продукта  
- data engineer будет собирать данные с форумов, сайтов, отделов компании и др., подготавливать их для аналитиков и DS-специалистов  


В зависимости от выделенного бюджета эти роли могут быть совмещены в нескольких специалистах.
